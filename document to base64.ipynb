```python
def pdf_to_base64(pdf_file_path):
    """Convert PDF file to Base64."""
    with open(pdf_file_path, "rb") as pdf_file:
        encoded_pdf = base64.b64encode(pdf_file.read())
        return encoded_pdf.decode('utf-8')

pdf_file_path = "D:\\Impelox\\document extraction\\samples\\Charbel OCR Docs\\Vehicle Possession  Certificate.pdf"

pdf_base64 = print(pdf_to_base64(pdf_file_path))
pdf_base64
```

```python
def image_to_base64(image_file_path):
    """Convert image file (JPG, PNG, etc.) to Base64."""
    with open(image_file_path, "rb") as image_file:
        encoded_image = base64.b64encode(image_file.read())
        return encoded_image.decode('utf-8')

image_file_path = "D:\\Impelox\\document extraction\\samples\\Charbel OCR Docs\\Vehicle Possession  Certificate_image.jpg"

image_base64 = print(image_to_base64(image_file_path))
image_base64
```
